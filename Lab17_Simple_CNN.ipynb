# Simple CNN
$$ OutputSize = \frac{InputSize - FilterSize + (2*Padding)}{Stride + 1} $$

```
import torch
import torch.nn as nn
```

```
conv = nn.Conv2d(1,1,7,stride=2,padding=0)
conv
```

```
inputs = torch.Tensor(1,1,64,64)
out = conv(inputs)
out.shape
```

```
conv = nn.Conv2d(1,1,5,stride=1,padding=0)
inputs = torch.Tensor(1,1,32,64)
print(conv(inputs).shape)
```

```
input  = torch.Tensor(1,1,28,28)
conv1 = nn.Conv2d(1,5,5)
pool = nn.MaxPool2d(2)
out = conv1(input)
print(out.size())
out2 = pool(out)
print(out2.size())
```
